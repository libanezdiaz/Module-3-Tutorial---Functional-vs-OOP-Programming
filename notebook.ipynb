This Binary search algorithm finds the position of a value that is in a sorted array. If the values appears more than one time, it will continue searching. If the number is not found, it returns -1. 

```python
def binarySearch(arr, k):
    left = 0
    right = len(arr) -1
    first_position = -1

    while left <= right:
        middle = (left + right) // 2

        if arr[middle] == k:
            first_position = middle
            right = middle - 1

        elif arr[middle] < k:
            left = middle + 1

        else:
            right = middle - 1
    return first_position

print(binarySearch([1,2,3,4,5],4))
print(binarySearch([11,22,33,44,55], 445))
print(binarySearch([1,1,1,1,2],1))


```
